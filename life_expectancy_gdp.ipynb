```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('all_data.csv')

df.head()
```

```python
df.shape
```

```python
print(df.Country.unique())
```

```python
print(df.Year.unique())
```

```python
df=df.rename({'Life expectancy at birth (years)':'LEAB(Y)'}, axis = 'columns')
df.head()
```

```python
plt.figure(figsize=(8,6))
sns.displot(df.GDP, rug = True, kde=False)
plt.xlabel('GDP in Trillions of U.S. Dollars')
```

```python
plt.figure(figsize=(8,6))
sns.displot(df.LEABY, rug = True, kde=False)
plt.xlabel('Life expectancy at birth (years)')
```

```python
dfMeans = df.drop('Year', axis = 1).groupby('Country').mean().reset_index()
dfMeans
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='LEABY', y='Country', data=dfMeans)
plt.xlabel('Average Life Expectancy at Birth (Years)')
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='GDP', y='Country', data=dfMeans)
plt.xlabel('Average GDP in Trillions of U.S. Dollars')
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=df.Year, y=df.GDP, hue=df.Country)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.ylabel('GDP in Trillions of U.S. Dollars')
plt.show()
plt.clf()
```

```python
graphGDP = sns.FacetGrid(df, col='Country', col_wrap=3, hue='Country', sharey = False)

graphGDP = (graphGDP.map(sns.lineplot,'Year','GDP').add_legend().set_axis_labels('Year', 'GDP in Trillions of U.S. Dolalrs'))
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=df.Year, y=df.LEABY, hue=df.Country)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.ylabel('Life Expectancy at Birth (Years)')
plt.show()
plt.clf()
```

```python
graphLEABY = sns.FacetGrid(df, col='Country', col_wrap=3, hue = 'Country', sharey = False)

graphLEABY = (graphLEABY.map(sns.lineplot, 'Year', 'LEABY').add_legend().set_axis_labels('Year', 'Life Expectancy at Birth (Years)'))

plt.show()
plt.clf()
```

```python
sns.scatterplot(x=df.LEABY, y=df.GDP, hue=df.Country).legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.show()
plt.clf()
```

```python
graph = sns.FacetGrid(df, col='Country', col_wrap=3, hue = 'Country', sharey = False)

graph = (graph.map(sns.scatterplot,'LEABY', 'GDP').add_legend().set_axis_labels('Life Expectancy at Birth (Years)', 'GDP in Trillions of U.S. Dollars'))

plt.show()
plt.clf()
```
